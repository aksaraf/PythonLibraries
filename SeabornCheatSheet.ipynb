**Setup**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

**Import Data**

```python
#we will use seaborn's built-in dataset
#names of datasets can be found by following code
print(sns.get_dataset_names())
```

```python
#we will use car crash dataset
crash_df = sns.load_dataset('car_crashes')
crash_df  
```

# Distribution Plots

**Distribution Plot**




```python
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)
#kde- kernel density estimation. It estimates the distribution of data.
```

**Joint Plot**

```python
# regression plot
sns.jointplot(x='alcohol', y='speeding', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='alcohol', y='speeding', data=crash_df, kind='kde')
```

**Hexagone distribution**

```python
sns.jointplot(x='alcohol', y='speeding', data=crash_df, kind='hex')
```

**KDE Plot**

```python
sns.kdeplot(crash_df['alcohol'], shade=True)
```

**Pair Plots**

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex', palette='Blues')
#hue separates data based on sex
```

**Rug Plot**

```python
sns.rugplot(tips_df['tip'])
```

**Styling**

```python
sns.set_style('white')
#we can also use darkgrid, whitegrid, dark, ticks

plt.figure(figsize=(10,6))

sns.set_context('paper', font_scale=1.4)
#we can also use talk, poster

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

sns.despine(left=False, bottom=False)
#for x & y axis to be shown or not
# True removes the axis line
```

# Categorical Plots

**Bar Plots**

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
# estimator default value is mean
#we can also use standard deviation(np.std), variance(np.var), covariance(np.cov)
#the small line refers to variance between the male/female and the total bill amount
```

**Count Plot**

```python
sns.countplot(x='sex', data=tips_df)
# counts the number of occurances
```

**Box Plot**

```python
plt.figure(figsize=(14,9))

sns.set_style('whitegrid')

sns.set_context('paper', font_scale=2) 

sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.legend(loc=2)
#to reposition the legend
#1 upper right
#2 upper left
#3 lower left
#4 lower right
```

**Volin Plot** 

```python
plt.figure(figsize=(14,9))

sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
# it uses kde estimation to plot  
#split=False will plot male female separately

plt.legend(loc=0)
```

**Strip Plot**

```python
plt.figure(figsize=(14,9))
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
#jitter spreads the data points so they don't stack one over the other
#dodge separates male and female data
plt.legend(loc=0)
```

**Swarm Plot**

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
#swarm plot is a combination of violin and strip plot
```

**Palettes**

```python
plt.figure(figsize=(8,6))

sns.set_style('white')

sns.set_context('talk')

sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='seismic')
#palette sets different coloring style
#more colors can be found on matplotlib
```

# Matrix Plots

**Heatmaps**

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

crash_mx = crash_df.corr()
#corelation function
# it tells how influential the varibale is on the result(target)

sns.heatmap(crash_mx, annot=True, cmap='Reds')
#annot adds numbers to the map
```

```python
#loading flights data
flights = sns.load_dataset('flights')
print(flights)
```

```python
# rearrange the data
flights = flights.pivot_table(index='month', columns='year', values='passengers')
print(flights)
```

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1)
```

**Cluster Map**

```python
sns.clustermap(flights, cmap='Blues', standard_scale=1)
```

**PairGrid**

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
iris_g.add_legend()
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species',
                     x_vars=["sepal_length", "sepal_width"],
                     y_vars=["petal_length", "petal_width"])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

**Facet Grid**

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3,
                       col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', "tip", edgecolor='w')

tips_fg.add_legend()
```

```python
#Define that we want to assign different markers to smokers and non smokers
tips_df = sns.load_dataset('tips')
kws = dict(s=50, linewidth =5)
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3,
                        hue_order=['Yes','No'], 
                        hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, "total_bill", "tip", **kws)
tips_fg.add_legend()
```

```python
att_df = sns.load_dataset('attention')
att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5,
                      height=1.5)
# col_wrap - number of charts per row
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
```
